# Исследование важных для детекции фич и сравнение с атакуемыми патчем

Гипотеза - на разных изображениях патч старается одинаково влиять на фичи - специально воздействуя именно на те, которые являлись наиболее важными для детекции целевого объекта на обучающей выборке

## Вспомогательные функции

```python
from __future__ import annotations

from dataclasses import dataclass
from typing import Any, Dict, List, Optional, Tuple

import math
import numpy as np
import torch
from PIL import Image

import matplotlib.pyplot as plt


# -------------------------
# Image conversions
# -------------------------

def pil_to_np_bgr(pil: Image.Image) -> np.ndarray:
    """PIL RGB -> np.uint8 BGR (H,W,3) for ultralytics predictor preprocess."""
    if pil.mode != "RGB":
        pil = pil.convert("RGB")
    rgb = np.asarray(pil)  # uint8 RGB
    bgr = rgb[..., ::-1].copy()
    return bgr


def pil_to_torch_rgb01(pil: Image.Image) -> torch.Tensor:
    """PIL -> torch float32 RGB in [0,1], shape [3,H,W] (CPU)."""
    if pil.mode != "RGB":
        pil = pil.convert("RGB")
    arr = np.asarray(pil).astype(np.float32) / 255.0
    t = torch.from_numpy(arr).permute(2, 0, 1).contiguous()
    return t


# -------------------------
# Patch application
# -------------------------

def apply_patch_to_image(
    base_pil: Image.Image,
    patch_pil: Image.Image,
    position_xy: Tuple[int, int] = (0, 0),
) -> Tuple[Image.Image, Optional[torch.Tensor], int]:
    """Paste patch onto base image at (x,y). Returns patched PIL, bbox_xyxy, area_px."""
    if base_pil.mode != "RGB":
        base_pil = base_pil.convert("RGB")
    if patch_pil.mode != "RGB":
        patch_pil = patch_pil.convert("RGB")
    base_w, base_h = base_pil.size
    patch_w, patch_h = patch_pil.size
    x, y = position_xy
    x1 = max(0, int(x))
    y1 = max(0, int(y))
    x2 = min(base_w, int(x) + patch_w)
    y2 = min(base_h, int(y) + patch_h)
    if x2 <= x1 or y2 <= y1:
        return base_pil.copy(), None, 0
    # Crop patch if it spills outside the base image.
    px1 = x1 - int(x)
    py1 = y1 - int(y)
    px2 = px1 + (x2 - x1)
    py2 = py1 + (y2 - y1)
    patch_crop = patch_pil.crop((px1, py1, px2, py2))
    out = base_pil.copy()
    out.paste(patch_crop, (x1, y1))
    bbox = torch.tensor([float(x1), float(y1), float(x2), float(y2)], dtype=torch.float32)
    area = int((x2 - x1) * (y2 - y1))
    return out, bbox, area


def letterbox_pil(pil: Image.Image, imgsz: int, color: Tuple[int, int, int] = (114, 114, 114)) -> Image.Image:
    """Resize+pad image to square imgsz, matching Ultralytics letterbox behavior."""
    if pil.mode != "RGB":
        pil = pil.convert("RGB")
    w, h = pil.size
    r = min(imgsz / h, imgsz / w)
    new_w = int(round(w * r))
    new_h = int(round(h * r))
    resized = pil.resize((new_w, new_h), Image.BILINEAR)
    out = Image.new("RGB", (imgsz, imgsz), color)
    pad_w = int((imgsz - new_w) / 2)
    pad_h = int((imgsz - new_h) / 2)
    out.paste(resized, (pad_w, pad_h))
    return out

from pathlib import Path
import random
from typing import List


def collect_images(
    folder: Path,
    n: int,
    extensions: tuple[str, ...] = (".png", ".jpg", ".jpeg"),
    shuffle: bool = False,
    seed: int | None = 42,
) -> List[str]:
    """
    Берёт n изображений из папки.
    Если shuffle=True — случайная выборка (с фиксируемым seed).
    """
    files = [
        p for p in folder.iterdir()
        if p.is_file() and p.suffix.lower() in extensions
    ]

    if not files:
        raise RuntimeError(f"В папке {folder} нет изображений")

    if shuffle:
        if seed is not None:
            random.seed(seed)
        random.shuffle(files)

    return [str(p) for p in files[:n]]
```

```python
N_SUCCESS = 10
N_FAIL = 10
SEED=17

successful_dir = Path("successful_examples")
unsuccessful_dir = Path("unsuccessful_examples")

image_paths = (
    collect_images(successful_dir, N_SUCCESS, shuffle=True, seed=SEED)
    + collect_images(unsuccessful_dir, N_FAIL, shuffle=True, seed=SEED)
)
patch_path = "data/patch.png"
```

## Конфигурация эксперимента

Здесь фиксируем параметры эксперимента, чтобы результаты были воспроизводимыми и сопоставимыми между изображениями.

**Что фиксируем:**
- `imgsz` и устройство.
- где накладываем патч: **после** letterbox (то есть на том же входе, который реально видит модель) или **до**.
- целевой класс (если задан): метрики успеха/важности считаются именно по нему.
- порог падения уверенности, по которому считаем атаку «успешной».

**Почему это важно:**
- Если менять препроцессинг/порог/цель, метрики будут несравнимы.


```python
@dataclass
class ExpConfig:
    # путь к весам (можно заменить на локальный)
    model_path: str = "yolo11s.pt"

    # размер входа
    imgsz: int = 640

    # устройство
    device: str = "cuda" if torch.cuda.is_available() else "cpu"

    # True: накладываем патч уже на letterbox-изображение (самый честный режим)
    # False: накладываем на оригинал, потом делаем letterbox
    apply_patch_after_letterbox: bool = True

    # позиция патча (x,y) в координатах того изображения, куда накладываем
    patch_xy: Tuple[int, int] = (0, 0)

    # целевой класс (None => берём max по всем классам)
    target_class_name: Optional[str] = "person"

    # успех атаки: падение скаляра уверенности >= threshold
    success_drop_thresh: float = 0.30

    # число каналов, используемое в top-k метриках и в абляциях
    topk_channels: int = 32

    # ROI scalar reduction for attribution: "max" | "logsumexp" | "topk_mean"
    roi_reduce: str = "logsumexp"

    # temperature for logsumexp (larger -> closer to max)
    roi_lse_temp: float = 20.0

    # k for topk_mean
    roi_topk: int = 10

        # --- ROI attribution layer selection ---
    # How many layers to keep for ROI-attribution (set 1 to mimic the previous "single good layer" behavior)
    roi_keep_topk_layers: int = 1

        # How many near-head conv layers to consider as candidates
    roi_candidate_last_n_layers: int = 8

    # How many examples from each group (success/fail) to use for scanning grad signal
    roi_scan_n_per_group: int = 2

    # Minimum median ma_abs.max required to accept a layer (0.0 keeps the best even if tiny)
    roi_min_grad_strength: float = 0.0




CFG = ExpConfig()

```

## Загрузка YOLO и определение скаляра уверенности `conf_pred`

Для проверки гипотезы нам нужен один скаляр `conf_pred`, который отражает «насколько модель уверена в наличии целевого объекта».

**Что считаем:**
- `conf_pred = max(confidence)` среди детекций целевого класса (или max среди всех классов, если `target_class_name=None`).

**Почему это подходит:**
- Это непосредственно согласовано с выходом детектора (после NMS) и удобно для бинарного разбиения на success/fail.

**Что это даст:**
- Строгую метку успеха атаки и скаляр, с которым можно соотносить важность фич.


```python
def _load_ultralytics_yolo(model_path: str):
    try:
        from ultralytics import YOLO
    except Exception as e:
        raise RuntimeError(
            "Не удалось импортировать ultralytics. Установите пакет ultralytics или убедитесь, что среда активна."
        ) from e
    return YOLO(model_path)


def _get_class_id_from_names(names: Dict[int, str], class_name: str) -> Optional[int]:
    class_name = class_name.lower().strip()
    for k, v in names.items():
        if str(v).lower().strip() == class_name:
            return int(k)
    return None


@torch.no_grad()
def yolo_predict_conf_scalar(
    yolo_model,
    pil_img: Image.Image,
    imgsz: int,
    target_class_id: Optional[int],
    conf: float = 0.001,
) -> Tuple[float, Any]:
    """Ultralytics predict -> scalar confidence (NMS-based) + raw Results."""
    bgr = pil_to_np_bgr(pil_img)
    res = yolo_model.predict(source=bgr, imgsz=imgsz, conf=conf, verbose=False)[0]

    if res.boxes is None or len(res.boxes) == 0:
        return 0.0, res

    confs = res.boxes.conf.detach().cpu().numpy()
    clss = res.boxes.cls.detach().cpu().numpy().astype(int)

    if target_class_id is None:
        return float(confs.max(initial=0.0)), res

    mask = (clss == int(target_class_id))
    if not np.any(mask):
        return 0.0, res

    return float(confs[mask].max(initial=0.0)), res

```

## Сбор clean/patched входов с контролем letterbox

Мы приводим изображение к тому виду, который реально видит модель: square `imgsz×imgsz` через letterbox.

**Два режима:**
- `apply_patch_after_letterbox=True`: патч накладывается прямо на letterbox-вход. Это наиболее корректно для анализа влияния на фичи модели.
- `False`: патч накладывается на оригинал, затем происходит letterbox. Это полезно как контрольный режим.

**Что это даст:**
- Контроль над тем, в каких координатах/масштабах патч воздействует на сеть.


```python
def build_clean_and_patched_letterboxed(
    base_pil: Image.Image,
    patch_pil: Image.Image,
    cfg: ExpConfig,
) -> Tuple[Image.Image, Image.Image, Optional[torch.Tensor]]:
    clean_lb = letterbox_pil(base_pil, imgsz=cfg.imgsz)

    if cfg.apply_patch_after_letterbox:
        patched_lb, bbox, _ = apply_patch_to_image(clean_lb, patch_pil, cfg.patch_xy)
        return clean_lb, patched_lb, bbox

    patched_orig, _bbox_orig, _ = apply_patch_to_image(base_pil, patch_pil, cfg.patch_xy)
    patched_lb = letterbox_pil(patched_orig, imgsz=cfg.imgsz)
    return clean_lb, patched_lb, None

```

## Выбор слоёв и снятие активаций (hooks)

Чтобы проверить «патч атакует одни и те же фичи» и «атакуемые фичи совпадают с важными», нам нужны:
- `F_l(clean)` и `F_l(patched)` для набора слоёв `l`
- `Δ_l = F_l(patched) - F_l(clean)`

**Почему hooks подходят:**
- Это минимально-инвазивный способ извлечь тензоры признаков из уже готовой реализации модели.

**Как выбираем слои:**
- Автоматически выбираем несколько `Conv2d` равномерно по глубине сети (backbone+neck+head смесь).
- При желании позже можно заменить на ручной список (например, конкретные уровни FPN/PAN).


```python
import torch.nn as nn


def pick_default_conv_layers(model_torch: nn.Module, n_layers: int = 12) -> List[str]:
    convs: List[str] = []
    for n, m in model_torch.named_modules():
        if isinstance(m, nn.Conv2d):
            convs.append(n)
    if not convs:
        raise RuntimeError("Не удалось найти Conv2d модули в модели.")

    idxs = np.linspace(0, len(convs) - 1, num=min(n_layers, len(convs))).round().astype(int)
    picked = []
    seen = set()
    for i in idxs:
        name = convs[int(i)]
        if name not in seen:
            picked.append(name)
            seen.add(name)
    return picked


def get_module_by_name(model_torch: nn.Module, name: str) -> nn.Module:
    d = dict(model_torch.named_modules())
    if name not in d:
        raise KeyError(f"Layer '{name}' not found.")
    return d[name]


@torch.no_grad()
def torch_preprocess_letterboxed(pil_img_lb: Image.Image) -> torch.Tensor:
    t = pil_to_torch_rgb01(pil_img_lb)
    return t.unsqueeze(0)  # [1,3,H,W]


def capture_activations_for_layers(
    model_torch: nn.Module,
    x_bchw: torch.Tensor,
    layer_names: List[str],
) -> Dict[str, torch.Tensor]:
    acts: Dict[str, torch.Tensor] = {}
    hooks = []

    def _mk_hook(name: str):
        def _hook(_m, _inp, out):
            if isinstance(out, torch.Tensor) and out.ndim == 4:
                acts[name] = out.detach().cpu()
        return _hook

    for ln in layer_names:
        m = get_module_by_name(model_torch, ln)
        hooks.append(m.register_forward_hook(_mk_hook(ln)))

    _ = model_torch(x_bchw)

    for h in hooks:
        h.remove()

    return acts

```

## Метрики похожести атакуемых фич (Δ) между изображениями

Первая часть гипотезы: патч на разных изображениях «бьёт» по похожим фичам.

Мы сравниваем `Δ_l` между изображениями несколькими комплементарными метриками:

- **cosine(vec(|Δ|))** — сравнение структуры изменения (инвариантность к масштабу, игнорируем знак).
- **Spearman по каналам** — совпадает ли ранжирование каналов по `mean|Δ|` (то есть «те же каналы»).
- **Jaccard@k по топ-k каналам** — интерпретируемая мера пересечения атакуемых каналов.
- **linear CKA** — стандартная метрика сходства репрезентаций, более устойчивая к линейным преобразованиям, чем L2.

**Что это даст:**
- Понимание, на каких слоях атака ведёт себя «стабильно» между сценами, а где становится контекстно-зависимой.


```python
def channel_energy(delta_bchw: torch.Tensor) -> np.ndarray:
    return delta_bchw.abs().mean(dim=(0, 2, 3)).numpy()


def spatial_energy(delta_bchw: torch.Tensor) -> np.ndarray:
    return delta_bchw.abs().sum(dim=1).squeeze(0).numpy()


def cosine_sim(a: np.ndarray, b: np.ndarray, eps: float = 1e-12) -> float:
    a = a.reshape(-1).astype(np.float64)
    b = b.reshape(-1).astype(np.float64)
    na = np.linalg.norm(a)
    nb = np.linalg.norm(b)
    if na < eps or nb < eps:
        return 0.0
    return float(np.dot(a, b) / (na * nb))


def spearman_corr(a: np.ndarray, b: np.ndarray) -> float:
    a = a.reshape(-1)
    b = b.reshape(-1)
    ra = a.argsort().argsort().astype(np.float64)
    rb = b.argsort().argsort().astype(np.float64)
    ra -= ra.mean()
    rb -= rb.mean()
    denom = np.linalg.norm(ra) * np.linalg.norm(rb)
    if denom < 1e-12:
        return 0.0
    return float(np.dot(ra, rb) / denom)


def jaccard_topk(a: np.ndarray, b: np.ndarray, k: int) -> float:
    k = int(min(k, a.size, b.size))
    ia = set(np.argsort(-a)[:k].tolist())
    ib = set(np.argsort(-b)[:k].tolist())
    inter = len(ia & ib)
    uni = len(ia | ib)
    return float(inter / max(1, uni))


def linear_cka(X: np.ndarray, Y: np.ndarray, eps: float = 1e-12) -> float:
    X = X.astype(np.float64)
    Y = Y.astype(np.float64)
    X = X - X.mean(axis=0, keepdims=True)
    Y = Y - Y.mean(axis=0, keepdims=True)
    XT_Y = X.T @ Y
    hsic = (XT_Y * XT_Y).sum()
    norm_x = (X.T @ X)
    norm_y = (Y.T @ Y)
    denom = math.sqrt((norm_x * norm_x).sum() * (norm_y * norm_y).sum())
    if denom < eps:
        return 0.0
    return float(hsic / denom)


def delta_similarity_metrics(delta1: torch.Tensor, delta2: torch.Tensor, topk: int) -> Dict[str, float]:
    v1 = delta1.abs().numpy().reshape(-1)
    v2 = delta2.abs().numpy().reshape(-1)
    cos = cosine_sim(v1, v2)

    d1 = channel_energy(delta1)
    d2 = channel_energy(delta2)
    spr = spearman_corr(d1, d2)
    jac = jaccard_topk(d1, d2, k=topk)

    _, C, H, W = delta1.shape
    X = delta1.squeeze(0).permute(1, 2, 0).reshape(H * W, C).abs().numpy()
    Y = delta2.squeeze(0).permute(1, 2, 0).reshape(H * W, C).abs().numpy()
    cka = linear_cka(X, Y)

    return {"cos_abs": cos, "spearman_chan": spr, "jaccard_topk": jac, "cka_abs": cka}

```

## Прогон: извлекаем `F_l(clean)`, `F_l(patched)` и `Δ_l` для всех изображений

Эта ячейка делает «логирование» всех нужных тензоров и скаляров:
- `conf_clean`, `conf_patch`, `drop`, `success`
- `acts_clean[l]`, `acts_patch[l]`, `deltas[l]`

**Почему так:**
- Далее мы будем многократно считать разные метрики по уже сохранённым тензорам без повторного прогона модели.


```python
# load model

yolo = _load_ultralytics_yolo(CFG.model_path)
model_torch: nn.Module = yolo.model
model_torch.to(CFG.device)
model_torch.eval()

names = getattr(yolo, "names", None)
if names is None:
    names = getattr(model_torch, "names", {})

if CFG.target_class_name is not None:
    target_class_id = _get_class_id_from_names(names, CFG.target_class_name)
    if target_class_id is None:
        print(f"[warn] target_class_name='{CFG.target_class_name}' не найден в model.names; использую max по всем классам")
else:
    target_class_id = None

layer_names = pick_default_conv_layers(model_torch, n_layers=12)
print("Selected layers:")
for ln in layer_names:
    print(" -", ln)

patch_pil = Image.open(patch_path).convert("RGB")

run_data: List[Dict[str, Any]] = []

for p in image_paths:
    base_pil = Image.open(p).convert("RGB")
    clean_lb, patched_lb, patch_bbox_lb = build_clean_and_patched_letterboxed(base_pil, patch_pil, CFG)

    conf_clean, res_clean = yolo_predict_conf_scalar(yolo, clean_lb, CFG.imgsz, target_class_id)
    conf_patch, res_patch = yolo_predict_conf_scalar(yolo, patched_lb, CFG.imgsz, target_class_id)

    x_clean = torch_preprocess_letterboxed(clean_lb).to(CFG.device)
    x_patch = torch_preprocess_letterboxed(patched_lb).to(CFG.device)

    acts_clean = capture_activations_for_layers(model_torch, x_clean, layer_names)
    acts_patch = capture_activations_for_layers(model_torch, x_patch, layer_names)

    deltas: Dict[str, torch.Tensor] = {}
    for ln in layer_names:
        if ln in acts_clean and ln in acts_patch:
            deltas[ln] = (acts_patch[ln] - acts_clean[ln])

    drop = float(conf_clean - conf_patch)
    success = bool(drop >= CFG.success_drop_thresh)

    run_data.append(
        {
            "path": p,
            "clean_lb": clean_lb,
            "patched_lb": patched_lb,
            "patch_bbox_lb": patch_bbox_lb,
            "conf_clean": float(conf_clean),
            "conf_patch": float(conf_patch),
            "drop": drop,
            "success": success,
            "acts_clean": acts_clean,
            "acts_patch": acts_patch,
            "deltas": deltas,
            "res_clean": res_clean,
            "res_patch": res_patch,
        }
    )

print("\nPer-image summary:")
for d in run_data:
    print(f"- {d['path']}: clean={d['conf_clean']:.3f} patched={d['conf_patch']:.3f} drop={d['drop']:.3f} success={d['success']}")

```

## Сходство `Δ` между изображениями по каждому слою

Здесь считаем метрики похожести `Δ_l` для всех пар изображений и агрегируем их по слоям.

**Что это даёт:**
- На каких уровнях сети патч действует консистентно между сценами.
- Где поведение становится зависимым от контекста.


```python
def compute_pairwise_delta_similarity(
    run_data: List[Dict[str, Any]],
    layer_names: List[str],
    topk: int,
) -> List[Dict[str, Any]]:
    rows: List[Dict[str, Any]] = []
    n = len(run_data)
    for i in range(n):
        for j in range(i + 1, n):
            di = run_data[i]["deltas"]
            dj = run_data[j]["deltas"]
            for ln in layer_names:
                if ln not in di or ln not in dj:
                    continue
                m = delta_similarity_metrics(di[ln], dj[ln], topk=topk)
                rows.append({"pair": f"{i}-{j}", "layer": ln, **m})
    return rows


pair_rows = compute_pairwise_delta_similarity(run_data, layer_names, topk=CFG.topk_channels)

if not pair_rows:
    print("Нужно >=2 изображения.")
else:
    by_layer: Dict[str, Dict[str, float]] = {}
    counts: Dict[str, int] = {}
    for r in pair_rows:
        ln = r["layer"]
        counts[ln] = counts.get(ln, 0) + 1
        if ln not in by_layer:
            by_layer[ln] = {"cos_abs": 0.0, "spearman_chan": 0.0, "jaccard_topk": 0.0, "cka_abs": 0.0}
        for k in by_layer[ln].keys():
            by_layer[ln][k] += float(r[k])

    print("\nΔ similarity (mean over pairs) by layer:")
    for ln in layer_names:
        if ln not in by_layer:
            continue
        c = counts[ln]
        m = {k: v / c for k, v in by_layer[ln].items()}
        print(f"- {ln}: cos={m['cos_abs']:.3f} spearman={m['spearman_chan']:.3f} jacc@k={m['jaccard_topk']:.3f} cka={m['cka_abs']:.3f}")

```

## Визуализация: spatial-карты `|Δ|` по слоям

Скалярные метрики важны, но без карт сложно интерпретировать, *где* именно меняются признаки.

Мы строим карты:
- `M_l(h,w) = sum_c |Δ_l(c,h,w)|`

и смотрим, как изменения:
- локализуются около патча
- «размазываются» по сцене

**Что это даёт:**
- быстрый визуальный признак того, насколько патч вызывает локальные vs глобальные изменения.


```python
def show_spatial_delta_maps(run_item: Dict[str, Any], layer_names: List[str], max_layers: int = 6):
    deltas = run_item["deltas"]
    to_show = [ln for ln in layer_names if ln in deltas][:max_layers]
    if not to_show:
        print("Нет deltas для визуализации.")
        return

    n = len(to_show)
    fig, axes = plt.subplots(2, n, figsize=(3 * n, 6))

    for j in range(n):
        axes[0, j].imshow(run_item["patched_lb"] if j > 0 else run_item["clean_lb"])
        axes[0, j].set_title("patched" if j > 0 else "clean")
        axes[0, j].axis("off")

    for j, ln in enumerate(to_show):
        m = spatial_energy(deltas[ln])
        axes[1, j].imshow(m)
        axes[1, j].set_title(f"|Δ| sum_c\n{ln}")
        axes[1, j].axis("off")

    plt.tight_layout()
    plt.show()


if run_data:
    show_spatial_delta_maps(run_data[0], layer_names, max_layers=6)

```

## Важные фичи для детекции: channel ablation в feature space

Теперь связываем «атакуемые фичи» с «важными фичами».

Мы используем **каузальный** тест важности каналов:
- на clean-изображении выбираем слой `l`
- зануляем один канал `c` в `F_l`
- считаем падение `conf_pred`

`importance(c) = conf_base - conf_ablated`

**Почему это сильный тест:**
- Он отвечает на причинный вопрос: *если убрать канал, упадёт ли детекция?*
- Он не зависит от качества градиентов и не требует backprop через NMS.

**Компромисс:**
- это дорого, поэтому оцениваем важность только на объединении:
  - топ-k атакуемых каналов (по `mean|Δ|`)
  - топ-k активных на clean (по `mean|F_clean|`)


```python
def _hook_zero_channels(ch_idxs: List[int]):
    ch_idxs = [int(x) for x in ch_idxs]

    def _hook(_m, _inp, out):
        if not (isinstance(out, torch.Tensor) and out.ndim == 4):
            return out
        out = out.clone()
        for c in ch_idxs:
            if 0 <= c < out.shape[1]:
                out[:, c, :, :] = 0.0
        return out

    return _hook


def pick_candidate_channels_for_layer(acts_clean: torch.Tensor, delta: torch.Tensor, topk: int) -> List[int]:
    d = delta.abs().mean(dim=(0, 2, 3)).numpy()
    attacked = np.argsort(-d)[:topk].tolist()

    a = acts_clean.abs().mean(dim=(0, 2, 3)).numpy()
    active = np.argsort(-a)[:topk].tolist()

    out: List[int] = []
    seen = set()
    for x in attacked + active:
        if int(x) not in seen:
            out.append(int(x))
            seen.add(int(x))
    return out


def compute_channel_importance_via_ablation(
    yolo_model,
    model_torch: nn.Module,
    clean_lb: Image.Image,
    layer_name: str,
    candidate_channels: List[int],
    cfg: ExpConfig,
    target_class_id: Optional[int],
) -> Dict[int, float]:
    conf_base, _ = yolo_predict_conf_scalar(yolo_model, clean_lb, cfg.imgsz, target_class_id)

    layer = get_module_by_name(model_torch, layer_name)
    imps: Dict[int, float] = {}

    for c in candidate_channels:
        # зануляем конкретный канал на выходе слоя
        h = layer.register_forward_hook(
            lambda _m, _inp, out, cc=int(c): _hook_zero_channels([cc])(_m, _inp, out)
        )
        try:
            conf_abl, _ = yolo_predict_conf_scalar(yolo_model, clean_lb, cfg.imgsz, target_class_id)
        finally:
            h.remove()
        imps[int(c)] = float(conf_base - conf_abl)

    return imps

```

## Метрики совпадения «атакуемые vs важные» и сравнение success/fail

Для каждого слоя и изображения мы имеем:
- `d_c = mean|Δ|` — насколько канал атакован патчем
- `imp_c` — насколько канал важен для детекции (абляция)

Дальше считаем совпадение:
- **Spearman(d, imp)** — совпадает ли ранжирование атакуемости и важности.
- **Jaccard@k(d, imp)** — пересечение топ-k атакуемых и топ-k важных.
- **cos(d, imp)** — простая непрерывная метрика согласованности.

**Что это даёт:**
- Прямую проверку гипотезы: в успешных атаках совпадение должно быть выше.


```python
alignment_rows: List[Dict[str, Any]] = []

for item in run_data:
    acts_clean = item["acts_clean"]
    deltas = item["deltas"]

    for ln in layer_names:
        if ln not in deltas or ln not in acts_clean:
            continue

        delta = deltas[ln]
        act = acts_clean[ln]

        cand = pick_candidate_channels_for_layer(act, delta, topk=CFG.topk_channels)

        d_full = channel_energy(delta)
        d_c = np.array([d_full[c] if c < d_full.size else 0.0 for c in cand], dtype=np.float64)

        imp_map = compute_channel_importance_via_ablation(
            yolo_model=yolo,
            model_torch=model_torch,
            clean_lb=item["clean_lb"],
            layer_name=ln,
            candidate_channels=cand,
            cfg=CFG,
            target_class_id=target_class_id,
        )
        imp_c = np.array([imp_map.get(int(c), 0.0) for c in cand], dtype=np.float64)

        alignment_rows.append(
            {
                "path": item["path"],
                "success": item["success"],
                "layer": ln,
                "drop": item["drop"],
                "spearman(d,imp)": spearman_corr(d_c, imp_c),
                "jaccard@k": jaccard_topk(d_c, imp_c, k=min(CFG.topk_channels, len(cand))),
                "cos(d,imp)": cosine_sim(d_c, imp_c),
                "n_channels": len(cand),
            }
        )


if not alignment_rows:
    print("alignment_rows пуст — проверьте, что hooks вернули активации.")
else:
    print("\nAlignment summary (mean over images) by layer and success:")
    for ln in layer_names:
        s = [r for r in alignment_rows if r["layer"] == ln and r["success"]]
        f = [r for r in alignment_rows if r["layer"] == ln and not r["success"]]
        if not s and not f:
            continue

        def _mean(rows, key):
            return float(np.mean([float(x[key]) for x in rows])) if rows else float("nan")

        print(
            f"- {ln}: spearman s={_mean(s,'spearman(d,imp)'):.3f} f={_mean(f,'spearman(d,imp)'):.3f} | "
            f"jacc s={_mean(s,'jaccard@k'):.3f} f={_mean(f,'jaccard@k'):.3f} | "
            f"cos s={_mean(s,'cos(d,imp)'):.3f} f={_mean(f,'cos(d,imp)'):.3f}"
        )
```

## Новые метрики для проверки гипотезы: L2 и более продвинутые альтернативы

Сейчас мы хотим проверить гипотезу сильнее, чем просто «пересечение top-каналов».

Ключевые идеи:

1) **Насколько сильно патч в среднем меняет признаки на слое** — это можно начать с простой L2 (точнее RMS) нормы `Δ_l`.

2) **Насколько сильно меняется само представление** между clean и patched:
- cosine similarity между `F_l(clean)` и `F_l(patched)` (чем ниже, тем сильнее изменение направления представления)
- linear CKA между `F_l(clean)` и `F_l(patched)` (устойчивее к масштабам и лучше отражает «подпространство»)

3) **Где пространственно проявляется изменение**:
- доля энергии `|Δ|` внутри области патча
- доля энергии `|Δ|` внутри области целевого объекта (мы берём bbox топ-детекции целевого класса на clean как прокси ROI)
- отношение «энергия в объекте / энергия в патче» — простой индикатор, насколько изменение «доходит» до объекта

4) **Насколько изменение локально или глобально (spread)**:
- энтропия распределения по spatial-карте `M_l(h,w) = sum_c |Δ|` (выше = более распределённо)

Мы посчитаем эти метрики для каждого изображения и слоя и сравним распределения для success vs fail.


```python
import pandas as pd


def _rms(t: torch.Tensor) -> float:
    """Root-mean-square for comparability across layers of different sizes."""
    x = t.detach().float()
    return float(torch.sqrt(torch.mean(x * x)).cpu().item())


def _cosine_vec(a: torch.Tensor, b: torch.Tensor, eps: float = 1e-12) -> float:
    av = a.detach().float().reshape(-1)
    bv = b.detach().float().reshape(-1)
    na = torch.linalg.norm(av)
    nb = torch.linalg.norm(bv)
    if float(na) < eps or float(nb) < eps:
        return float("nan")
    return float((av @ bv / (na * nb)).cpu().item())


def _cka_clean_patched(act_clean: torch.Tensor, act_patch: torch.Tensor) -> float:
    """Linear CKA between clean and patched activations on a layer."""
    # act: [1,C,H,W]
    _, C, H, W = act_clean.shape
    X = act_clean.squeeze(0).permute(1, 2, 0).reshape(H * W, C).cpu().numpy()
    Y = act_patch.squeeze(0).permute(1, 2, 0).reshape(H * W, C).cpu().numpy()
    return float(linear_cka(X, Y))


def _bbox_to_fm_coords(bbox_xyxy: np.ndarray, H: int, W: int, imgsz: int) -> Tuple[int, int, int, int]:
    """Map bbox in letterbox pixel coords (imgsz×imgsz) to feature-map coords (H×W)."""
    x1, y1, x2, y2 = [float(x) for x in bbox_xyxy]
    fx1 = int(np.floor(x1 / imgsz * W))
    fx2 = int(np.ceil(x2 / imgsz * W))
    fy1 = int(np.floor(y1 / imgsz * H))
    fy2 = int(np.ceil(y2 / imgsz * H))
    fx1 = int(np.clip(fx1, 0, W))
    fx2 = int(np.clip(fx2, 0, W))
    fy1 = int(np.clip(fy1, 0, H))
    fy2 = int(np.clip(fy2, 0, H))
    if fx2 <= fx1 or fy2 <= fy1:
        return 0, 0, 0, 0
    return fx1, fy1, fx2, fy2


def _spatial_energy_map(delta: torch.Tensor) -> np.ndarray:
    # delta: [1,C,H,W] -> map [H,W]
    return spatial_energy(delta)


def _entropy_map(m: np.ndarray, eps: float = 1e-12) -> float:
    v = m.astype(np.float64)
    s = v.sum()
    if s < eps:
        return float("nan")
    p = v.reshape(-1) / s
    p = np.clip(p, eps, 1.0)
    return float(-(p * np.log(p)).sum())


def _get_target_bbox_from_results(res, target_class_id: Optional[int]) -> Optional[np.ndarray]:
    """Use top detection bbox on clean as proxy ROI for target object."""
    if res is None or res.boxes is None or len(res.boxes) == 0:
        return None
    boxes = res.boxes
    confs = boxes.conf.detach().cpu().numpy()
    clss = boxes.cls.detach().cpu().numpy().astype(int)
    xyxy = boxes.xyxy.detach().cpu().numpy()

    if target_class_id is None:
        idx = int(np.argmax(confs))
        return xyxy[idx]

    mask = (clss == int(target_class_id))
    if not np.any(mask):
        return None
    idxs = np.where(mask)[0]
    idx = int(idxs[np.argmax(confs[mask])])
    return xyxy[idx]

```

## Подсчёт метрик по каждому изображению и слою

Мы построим таблицу, где каждая строка — (image, layer), а колонки — метрики.

**Метрики по слою:**
- `delta_rms`: RMS-норма `Δ_l`
- `cos_clean_patch`: cosine(F_clean, F_patch)
- `cka_clean_patch`: linear CKA(F_clean, F_patch)
- `patch_frac`: доля энергии `|Δ|` внутри bbox патча
- `obj_frac`: доля энергии `|Δ|` внутри bbox целевого объекта (прокси по top-detection на clean)
- `obj_over_patch`: отношение `obj_energy / patch_energy`
- `entropy`: энтропия spatial-карты `sum_c |Δ|` (мера «распространённости»)

**Почему это может разделить success/fail:**
- Для fail часто ожидается, что изменения остаются локальными (высокий patch_frac, низкий obj_frac, низкий obj_over_patch).
- Для success ожидается больший перенос изменений в область объекта и/или больший глобальный spread.


```python
rows = []

for item in run_data:
    # proxy bbox объекта по clean results
    obj_bbox = _get_target_bbox_from_results(item.get("res_clean", None), target_class_id)
    patch_bbox = item.get("patch_bbox_lb", None)
    if isinstance(patch_bbox, torch.Tensor):
        patch_bbox = patch_bbox.detach().cpu().numpy()

    for ln in layer_names:
        acts_clean = item["acts_clean"].get(ln, None)
        acts_patch = item["acts_patch"].get(ln, None)
        delta = item["deltas"].get(ln, None)
        if acts_clean is None or acts_patch is None or delta is None:
            continue

        # base metrics
        delta_rms = _rms(delta)
        cos_cp = _cosine_vec(acts_clean, acts_patch)
        cka_cp = _cka_clean_patched(acts_clean, acts_patch)

        # spatial map
        m = _spatial_energy_map(delta)
        H, W = m.shape
        tot = float(np.sum(m))

        patch_frac = float("nan")
        obj_frac = float("nan")
        obj_over_patch = float("nan")

        if patch_bbox is not None:
            fx1, fy1, fx2, fy2 = _bbox_to_fm_coords(patch_bbox, H, W, CFG.imgsz)
            if fx2 > fx1 and fy2 > fy1 and tot > 0:
                patch_e = float(np.sum(m[fy1:fy2, fx1:fx2]))
                patch_frac = patch_e / tot
            else:
                patch_e = float("nan")
        else:
            patch_e = float("nan")

        if obj_bbox is not None:
            ox1, oy1, ox2, oy2 = _bbox_to_fm_coords(obj_bbox, H, W, CFG.imgsz)
            if ox2 > ox1 and oy2 > oy1 and tot > 0:
                obj_e = float(np.sum(m[oy1:oy2, ox1:ox2]))
                obj_frac = obj_e / tot
            else:
                obj_e = float("nan")
        else:
            obj_e = float("nan")

        if (not np.isnan(obj_e)) and (not np.isnan(patch_e)) and patch_e > 1e-12:
            obj_over_patch = obj_e / patch_e

        ent = _entropy_map(m)

        rows.append(
            {
                "path": item["path"],
                "success": bool(item["success"]),
                "drop": float(item["drop"]),
                "layer": ln,
                "delta_rms": delta_rms,
                "cos_clean_patch": cos_cp,
                "cka_clean_patch": cka_cp,
                "patch_frac": patch_frac,
                "obj_frac": obj_frac,
                "obj_over_patch": obj_over_patch,
                "entropy": ent,
            }
        )


df = pd.DataFrame(rows)
print("rows:", len(df))
df.head()

```

## Агрегация по success/fail и оценка эффекта (Cohen's d + permutation p-value)

Для каждой метрики и каждого слоя сравним распределения в группах success vs fail.

Мы выводим:
- средние по группам
- Cohen’s d (размер эффекта)
- permutation p-value (устойчиво при малых выборках)

**Интерпретация:**
- если гипотеза верна, ожидаем, что хотя бы на части слоёв метрики типа `obj_frac`/`obj_over_patch`/`entropy` будут заметно выше (или ниже) в success.


```python
from collections import defaultdict


def cohens_d(x: np.ndarray, y: np.ndarray) -> float:
    x = x[~np.isnan(x)]
    y = y[~np.isnan(y)]
    if len(x) < 2 or len(y) < 2:
        return float("nan")
    mx, my = float(np.mean(x)), float(np.mean(y))
    vx, vy = float(np.var(x, ddof=1)), float(np.var(y, ddof=1))
    sp = math.sqrt(((len(x) - 1) * vx + (len(y) - 1) * vy) / max(1, (len(x) + len(y) - 2)))
    if sp < 1e-12:
        return float("nan")
    return (mx - my) / sp


def perm_pvalue_mean_diff(x: np.ndarray, y: np.ndarray, n_perm: int = 3000, seed: int = 0) -> float:
    x = x[~np.isnan(x)]
    y = y[~np.isnan(y)]
    if len(x) == 0 or len(y) == 0:
        return float("nan")
    rng = np.random.default_rng(seed)
    obs = float(np.mean(x) - np.mean(y))
    pooled = np.concatenate([x, y])
    n = len(x)
    cnt = 0
    for _ in range(n_perm):
        rng.shuffle(pooled)
        x2 = pooled[:n]
        y2 = pooled[n:]
        diff = float(np.mean(x2) - np.mean(y2))
        if abs(diff) >= abs(obs):
            cnt += 1
    return (cnt + 1) / (n_perm + 1)


metrics = ["delta_rms", "cos_clean_patch", "cka_clean_patch", "patch_frac", "obj_frac", "obj_over_patch", "entropy"]

summary_rows = []
for ln in layer_names:
    df_l = df[df["layer"] == ln]
    if len(df_l) == 0:
        continue
    for m in metrics:
        s = df_l[df_l["success"]][m].to_numpy(dtype=np.float64)
        f = df_l[~df_l["success"]][m].to_numpy(dtype=np.float64)
        ms = float(np.nanmean(s)) if np.any(~np.isnan(s)) else float("nan")
        mf = float(np.nanmean(f)) if np.any(~np.isnan(f)) else float("nan")
        dval = cohens_d(s, f)
        pval = perm_pvalue_mean_diff(s, f, n_perm=2000, seed=42)
        summary_rows.append({"layer": ln, "metric": m, "mean_success": ms, "mean_fail": mf, "cohens_d": dval, "perm_p": pval})

summary = pd.DataFrame(summary_rows)

# Покажем топ-результаты по |d|
summary["abs_d"] = summary["cohens_d"].abs()
summary_sorted = summary.sort_values(["abs_d", "perm_p"], ascending=[False, True])

print("Top signals by effect size (|Cohen's d|):")
display(summary_sorted.head(25))

```

## Сводка по метрикам по всей сети (агрегация по слоям)

Иногда удобнее иметь один скаляр на изображение.
Соберём агрегированные метрики, усредняя по слоям (или по последним слоям), и посмотрим, что лучше разделяет success/fail.

**Зачем:**
- если один слой даёт шум, агрегат может быть устойчивее
- можно напрямую сравнивать разные метрики как «скор» для прогнозирования успеха


```python
# агрегируем по всем слоям
agg_rows = []
for (path, success), g in df.groupby(["path", "success"]):
    row = {"path": path, "success": bool(success)}
    for m in metrics:
        row[f"{m}_mean_all"] = float(np.nanmean(g[m].to_numpy(dtype=np.float64)))
    agg_rows.append(row)

agg = pd.DataFrame(agg_rows)

# печать простого сравнения (mean + effect size)
print("\nAggregated metrics (mean over layers):")
for m in metrics:
    x = agg[agg["success"]][f"{m}_mean_all"].to_numpy(dtype=np.float64)
    y = agg[~agg["success"]][f"{m}_mean_all"].to_numpy(dtype=np.float64)
    print(
        f"- {m}: mean_s={np.nanmean(x):.4f} mean_f={np.nanmean(y):.4f} d={cohens_d(x,y):.3f} p={perm_pvalue_mean_diff(x,y, n_perm=3000, seed=0):.3f}"
    )

# опционально: посмотрим корреляцию метрик с drop
print("\nCorrelation with drop (Spearman-like rank corr):")

def rank_corr(x: np.ndarray, y: np.ndarray) -> float:
    x = x.astype(np.float64)
    y = y.astype(np.float64)
    mask = (~np.isnan(x)) & (~np.isnan(y))
    x = x[mask]
    y = y[mask]
    if len(x) < 3:
        return float("nan")
    rx = x.argsort().argsort().astype(np.float64)
    ry = y.argsort().argsort().astype(np.float64)
    rx -= rx.mean(); ry -= ry.mean()
    denom = np.linalg.norm(rx) * np.linalg.norm(ry)
    if denom < 1e-12:
        return float("nan")
    return float((rx @ ry) / denom)

# join drop
drop_map = {d["path"]: float(d["drop"]) for d in run_data}
agg["drop"] = agg["path"].map(drop_map)

for m in metrics:
    r = rank_corr(agg[f"{m}_mean_all"].to_numpy(dtype=np.float64), agg["drop"].to_numpy(dtype=np.float64))
    print(f"- {m}: rank_corr={r:.3f}")

```

## ROI-gated важность фич на слое и совпадение с изменениями от патча (Grad*Act)

Твой исходный вопрос формулируется так:

1) **Можно ли сказать:** «на слое `l` вот такой набор фич сильнее всего помогает предсказать целевой объект»?
2) **И проверяем ли мы:** «патч меняет именно эти важные фичи»?

До этого момента мы в основном:
- измеряли **что меняет патч** (`Δ_l`) и как это стабильно между изображениями,
- а также измеряли **куда уходит энергия изменений** (в ROI объекта vs в ROI патча).

Здесь добавляем прямую связь «важность ↔ изменение»:

### Важность фич на слое
Используем **Grad*Act** по feature map слоя:
- берём дифференцируемый скаляр `S_raw` из сырых предсказаний головы (до NMS)
- считаем градиент `∂S_raw/∂F_l`
- строим карту важности `A_l = |grad * act|`

Это даёт:
- **важность каналов**: `a_c = mean_{h,w} A_l[c,h,w]`
- **важность по пространству**: `a(h,w) = sum_c A_l[c,h,w]`

### Совпадение важного с тем, что меняет патч
Для каждого слоя считаем spatial-карты:
- `MΔ(h,w) = sum_c |Δ_l|`
- `MA(h,w) = sum_c |grad*act|`

И сравниваем их:
- глобально: cosine(MΔ, MA)
- **в ROI объекта**: cosine(MΔ_ROI, MA_ROI)
- доля важности в ROI объекта: `attr_obj_frac`

Если твоя гипотеза верна, то для success ожидаем выше:
- `attr_obj_frac`
- `cos_roi` (совпадение именно там, где объект)


```python
import contextlib


def _extract_pred_tensor(model_out: Any) -> torch.Tensor:
    """Best-effort: extract prediction tensor from Ultralytics model forward output."""
    if isinstance(model_out, torch.Tensor):
        return model_out

    if isinstance(model_out, (list, tuple)):
        for obj in model_out:
            if isinstance(obj, torch.Tensor) and obj.ndim >= 3:
                return obj
        tensors = [o for o in model_out if isinstance(o, torch.Tensor)]
        if tensors:
            return max(tensors, key=lambda t: t.numel())

    if isinstance(model_out, dict):
        for v in model_out.values():
            if isinstance(v, torch.Tensor) and v.ndim >= 3:
                return v

    raise RuntimeError(f"Cannot extract pred tensor from output type: {type(model_out)}")


def _to_bnd(pred: torch.Tensor) -> torch.Tensor:
    """Convert prediction to shape [B, N, D] when possible."""
    if pred.ndim != 3:
        return pred

    B, A, B_or_D = pred.shape

    # Heuristic: if last dim is huge (>>256), it's likely N, so format is [B, D, N]
    if B_or_D > 256 and A <= 256:
        return pred.permute(0, 2, 1).contiguous()  # [B, N, D]

    return pred


def raw_conf_scalar_from_pred(pred_bnd: torch.Tensor, target_class_id: Optional[int]) -> torch.Tensor:
    """Differentiable confidence-like scalar from pred [B,N,D]."""
    if pred_bnd.ndim != 3:
        return pred_bnd.abs().max()

    B, N, D = pred_bnd.shape

    # Common case: [x,y,w,h,obj,cls...]
    if D >= 6:
        obj = pred_bnd[..., 4].sigmoid()
        cls = pred_bnd[..., 5:].sigmoid()
        if target_class_id is None:
            score = (obj * cls.max(dim=-1).values).max()
        else:
            t = int(target_class_id)
            if t < cls.shape[-1]:
                score = (obj * cls[..., t]).max()
            else:
                score = obj.max()
        return score

    # Fallback: treat remaining dims as class logits
    if D > 4:
        cls = pred_bnd[..., 4:].sigmoid()
        if target_class_id is None:
            return cls.max()
        t = int(target_class_id)
        if t < cls.shape[-1]:
            return cls[..., t].max()
        return cls.max()

    return pred_bnd.abs().max()


def grad_act_importance_maps(
    model_torch: nn.Module,
    x_bchw: torch.Tensor,
    layer_name: str,
    target_class_id: Optional[int],
    obj_bbox_xyxy: Optional[np.ndarray] = None,
) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """Return (channel_importance[C], spatial_importance[H,W]) using |grad*act| on the layer."""
    layer = get_module_by_name(model_torch, layer_name)

    store: Dict[str, torch.Tensor] = {}

    # Найдём detect-head (последний модуль класса Detect) — у него есть nc/reg_max и training-ветка без _inference.
    detect_head = None
    for m in model_torch.modules():
        if m.__class__.__name__ == "Detect":
            detect_head = m
    if detect_head is None:
        raise RuntimeError("Cannot find Ultralytics Detect head in model_torch.modules()")

    nc = int(getattr(detect_head, "nc", 0))
    if nc <= 0:
        raise RuntimeError("Detect head has invalid nc; cannot compute raw cls scalar")

    def _roi_slice_for_level(H: int, W: int) -> Tuple[slice, slice]:
        # ROI bbox задан в координатах letterbox (imgsz×imgsz)
        # Переводим в координаты текущей feature map (H×W)
        if obj_bbox_xyxy is None:
            return slice(0, H), slice(0, W)

        x1, y1, x2, y2 = [float(v) for v in obj_bbox_xyxy]
        fx1 = int(np.floor(x1 / CFG.imgsz * W))
        fx2 = int(np.ceil(x2 / CFG.imgsz * W))
        fy1 = int(np.floor(y1 / CFG.imgsz * H))
        fy2 = int(np.ceil(y2 / CFG.imgsz * H))

        fx1 = int(np.clip(fx1, 0, W))
        fx2 = int(np.clip(fx2, 0, W))
        fy1 = int(np.clip(fy1, 0, H))
        fy2 = int(np.clip(fy2, 0, H))

        # если ROI выродился — fallback на full map
        if fx2 <= fx1 or fy2 <= fy1:
            return slice(0, H), slice(0, W)

        return slice(fy1, fy2), slice(fx1, fx2)

    def _hook(_m, _inp, out):
        # В Ultralytics дальше по графу часто есть inplace-активации (SiLU(inplace=True) и т.п.).
        # Нельзя отдавать вниз leaf-тензор с requires_grad=True, иначе inplace упадёт.
        # Решение: создаём leaf out_leaf для градиента, но вниз отдаём НЕ-leaf out_nonleaf.
        if isinstance(out, torch.Tensor) and out.ndim == 4:
                        # Важно: приводим к float32, иначе grad*act может уйти в нули из-за FP16 underflow.
            out_leaf = out.detach().to(torch.float32).clone()
            out_leaf.requires_grad_(True)
            store["act"] = out_leaf
            out_nonleaf = out_leaf * 1.0  # делает non-leaf, безопасно для inplace дальше
            return out_nonleaf
        return out

    h = layer.register_forward_hook(_hook)
    try:
        model_torch.zero_grad(set_to_none=True)
        with torch.enable_grad():
            # КРИТИЧНО: избегаем Detect._inference(), потому что он может использовать кэшированные inference-тензоры
            # (anchors/shape/strides), полученные во время предыдущих predict() вызовов, и autograd на них падает.
            # Поэтому временно переключаем Detect-head в training-ветку: она возвращает raw карты до decode.
            prev_train_flag = detect_head.training
            detect_head.train(True)
            try:
                if hasattr(model_torch, "_predict_once"):
                    out = model_torch._predict_once(x_bchw)
                else:
                    raise RuntimeError("Ultralytics model has no _predict_once")
            finally:
                detect_head.train(prev_train_flag)

            # out в training-ветке Detect обычно list[Tensor] по уровням, shape [B, (reg_max*4+nc), H, W]
            # Нам нужен дифференцируемый скаляр, отражающий уверенность по целевому классу.
            if isinstance(out, (list, tuple)) and len(out) > 0 and isinstance(out[0], torch.Tensor) and out[0].ndim == 4:
                scores = []
                for xi in out:
                    cls = xi[:, -nc:, :, :]  # [B,nc,H,W]
                    _, _, H, W = cls.shape
                    ys, xs = _roi_slice_for_level(H, W)

 # выбираем карту скоров внутри ROI
                    if target_class_id is None:
                        roi_scores = cls[:, :, ys, xs]  # [B,nc,roiH,roiW]
                    else:
                        t = int(target_class_id)
                        if 0 <= t < nc:
                            roi_scores = cls[:, t:t+1, ys, xs]  # [B,1,roiH,roiW]
                        else:
                            roi_scores = cls[:, :, ys, xs]

                    # редукция по ROI: max / logsumexp / top-k mean
                    if CFG.roi_reduce == "max":
                        scores.append(roi_scores.max())
                    elif CFG.roi_reduce == "logsumexp":
                        flat = roi_scores.reshape(roi_scores.shape[0], -1)
                        lse = torch.logsumexp(flat * float(CFG.roi_lse_temp), dim=1) / float(CFG.roi_lse_temp)
                        scores.append(lse.max())
                    elif CFG.roi_reduce == "topk_mean":
                        flat = roi_scores.reshape(roi_scores.shape[0], -1)
                        k = int(min(int(CFG.roi_topk), flat.shape[1]))
                        k = max(1,k) # ???
                        topk = torch.topk(flat, k=k, dim=1).values
                        scores.append(topk.mean())
                    else:
                        raise RuntimeError(f"Unknown CFG.roi_reduce={CFG.roi_reduce}")

                s = torch.stack(scores).max()
            else:
                # fallback: если формат другой, пытаемся старым способом
                pred = _extract_pred_tensor(out)
                pred = _to_bnd(pred)
                s = raw_conf_scalar_from_pred(pred, target_class_id)

            s.backward()
            act_tmp = store.get("act", None)
            if act_tmp is not None and act_tmp.grad is not None:
                gmax = float(act_tmp.grad.detach().abs().max().cpu().item())
                if gmax == 0.0:
                    print(f"[warn] zero grad captured at layer={layer_name} | roi_reduce={CFG.roi_reduce} | s={float(s.detach().cpu().item()):.4f}")

            act = store.get("act", None)
            if act is None or act.grad is None:
                raise RuntimeError("Grad*Act failed: activation/grad not captured")

             # Важно: считаем в float32, иначе при fp16/amp возможен underflow -> карта станет нулевой.
            ga_signed = (act.grad.to(torch.float32) * act.to(torch.float32))  # [1,C,H,W], signed
            ga_abs = ga_signed.abs()  # [1,C,H,W], non-negative

            ch = ga_abs.mean(dim=(0, 2, 3)).detach().cpu().numpy()
            sp_abs = ga_abs.sum(dim=1).squeeze(0).detach().cpu().numpy()       # [H,W]
            sp_signed = ga_signed.sum(dim=1).squeeze(0).detach().cpu().numpy() # [H,W]

            return ch, sp_abs, sp_signed
    finally:
        h.remove()


def _cosine_map(a: np.ndarray, b: np.ndarray, eps: float = 1e-12) -> float:
    a = a.reshape(-1).astype(np.float64)
    b = b.reshape(-1).astype(np.float64)
    na = np.linalg.norm(a)
    nb = np.linalg.norm(b)
    if na < eps or nb < eps:
        return float("nan")
    return float((a @ b) / (na * nb))


def _safe_sum(m: np.ndarray) -> float:
    s = float(np.sum(m))
    return s

```

## Подсчёт ROI-gated метрик «важность ↔ Δ» и сравнение success/fail

Мы считаем метрики только на последних слоях (вблизи головы), потому что:
- это быстрее
- именно там ожидаем причинную связь с итоговым скором

Для каждого (image, layer):
- `attr_obj_frac`: доля важности MA в ROI объекта
- `cos_global`: cosine(MΔ, MA)
- `cos_roi`: cosine(MΔ_ROI, MA_ROI)

Затем агрегируем по слоям до «один скаляр на изображение» и сравниваем success vs fail.


```python
#
# выберем кандидатов слоёв для attribution.
# Важно: cls-скаляр зависит в первую очередь от классификационной ветки head (обычно cv3),
# тогда как reg-ветка (cv2) может давать строго нулевой градиент.
# Поэтому:
# 1) берём near-head слои (последние N из layer_names)
# 2) фильтруем по имени: исключаем dfl и (по умолчанию) отдаём приоритет cv3

cand_n = int(getattr(CFG, "roi_candidate_last_n_layers", 8))
near_head = [ln for ln in layer_names if "dfl" not in ln][-cand_n:]
prefer_cls = [ln for ln in near_head if ".cv3" in ln]
other = [ln for ln in near_head if ".cv3" not in ln]
roi_layer_candidates = prefer_cls + other  # cv3 first, but keep others

print("ROI layer candidates:")
for ln in roi_layer_candidates:
    print(" -", ln)


def _layer_grad_strength(example_item: Dict[str, Any], ln: str) -> float:
    """Return ma_abs.max for a single (image, layer). If errors -> 0."""
    try:
        obj_bbox = _get_target_bbox_from_results(example_item.get("res_clean", None), target_class_id)
        if obj_bbox is None:
            return 0.0
        x_clean = torch_preprocess_letterboxed(example_item["clean_lb"]).to(CFG.device)
        _ch, ma_abs, _ma_signed = grad_act_importance_maps(
            model_torch, x_clean, ln, target_class_id, obj_bbox_xyxy=obj_bbox
        )
        return float(np.nanmax(ma_abs))
    except Exception as e:
        print(f"[roi-layer-scan] layer={ln} ERROR: {type(e).__name__}: {e}")
        return 0.0


# Быстрый скан: проверяем градиент на небольшом подмножестве (2 success + 2 fail, если есть)
# Build scan set: a few success and a few fail images
scan_k = int(getattr(CFG, "roi_scan_n_per_group", 2))
scan_items: List[Dict[str, Any]] = []
scan_items += [it for it in run_data if it.get("success", False)][:scan_k]
scan_items += [it for it in run_data if not it.get("success", False)][:scan_k]
if len(scan_items) == 0:
    scan_items = run_data[: min(2 * scan_k, len(run_data))]

layer_scores: Dict[str, float] = {}
for ln in roi_layer_candidates:
    vals = [_layer_grad_strength(it, ln) for it in scan_items]
    layer_scores[ln] = float(np.median(vals)) if len(vals) else 0.0

print("\nROI candidate grad strength (median ma_abs.max over scan set):")
for ln, sc in sorted(layer_scores.items(), key=lambda kv: -kv[1]):
    print(f" - {ln}: {sc:.3e}")

# Pick top-K layers by score, with an optional minimum threshold
keep_k = int(getattr(CFG, "roi_keep_topk_layers", 1))
min_sc = float(getattr(CFG, "roi_min_grad_strength", 0.0))

sorted_layers = [ln for ln, _ in sorted(layer_scores.items(), key=lambda kv: -kv[1])]
roi_layers = [ln for ln in sorted_layers if layer_scores.get(ln, 0.0) > min_sc][:keep_k]

# Fallback: if everything is <= min_sc, keep the best-scoring layer anyway
if len(roi_layers) == 0 and len(sorted_layers) > 0:
    roi_layers = [sorted_layers[0]]

print("\nROI layers used for attribution (top-K best):")
for ln in roi_layers:
    print(f" - {ln} (score={layer_scores.get(ln,0.0):.3e})")

```

## Повторная проверка: похожи ли Δ-фичи между картинками на выбранных «хороших» слоях

Ранее мы считали `Δ similarity (mean over pairs)` по фиксированному списку слоёв `layer_names`.
Теперь мы хотим посмотреть то же самое, но **только на выбранных слоях `roi_layers`**, для которых:
- Grad*Act карта важности ненулевая и стабильно считается
- эти слои реально лежат на пути к нашему object-conditioned ROI-скаляру

Это важно, потому что:
- некоторые near-head слои (например reg-ветка) дают нулевой grad и могут быть нерелевантны для cls-скаляра
- анализ Δ на «хороших» слоях должен быть более интерпретируемым

Метрики (те же, что и раньше):
- `cos_abs`: cosine(vec(|Δ|))
- `spearman_chan`: Spearman по каналам (по mean|Δ|)
- `jaccard_topk`: пересечение top-k атакуемых каналов
- `cka_abs`: linear CKA по |Δ| (samples=H·W, features=C)


```python
# Sanity: ensure roi_layers exists
if "roi_layers" not in globals() or not roi_layers:
    print("[warn] roi_layers is empty; falling back to last layer in layer_names")
    selected_layers_for_delta = [layer_names[-2]]
else:
    selected_layers_for_delta = list(roi_layers)

print("Selected layers for Δ similarity:")
for ln in selected_layers_for_delta:
    print(" -", ln)

# Compute pairwise similarity rows restricted to selected layers
pair_rows_roi = compute_pairwise_delta_similarity(run_data, selected_layers_for_delta, topk=CFG.topk_channels)

if not pair_rows_roi:
    print("Need >=2 images with deltas on these layers.")
else:
    by_layer = {}
    counts = {}
    for r in pair_rows_roi:
        ln = r["layer"]
        counts[ln] = counts.get(ln, 0) + 1
        if ln not in by_layer:
            by_layer[ln] = {"cos_abs": 0.0, "spearman_chan": 0.0, "jaccard_topk": 0.0, "cka_abs": 0.0}
        for k in by_layer[ln].keys():
            by_layer[ln][k] += float(r[k])

    print("\nΔ similarity (mean over pairs) on selected layers:")
    for ln in selected_layers_for_delta:
        if ln not in by_layer:
            continue
        c = counts[ln]
        m = {k: v / c for k, v in by_layer[ln].items()}
        print(f"- {ln}: cos={m['cos_abs']:.3f} spearman={m['spearman_chan']:.3f} jacc@k={m['jaccard_topk']:.3f} cka={m['cka_abs']:.3f} | n_pairs={c}")

# Optional: compare within success group vs within fail group (uses image-level success label)
# This helps see whether the patch-induced Δ patterns are more consistent inside one regime.

def _subset_pair_rows(items, layers):
    return compute_pairwise_delta_similarity(items, layers, topk=CFG.topk_channels)

succ_items = [it for it in run_data if it.get("success", False)]
fail_items = [it for it in run_data if not it.get("success", False)]


for tag, subset in [("success", succ_items), ("fail", fail_items)]:
    if len(subset) < 2:
        print(f"\nΔ similarity on selected layers within {tag}: not enough images (n={len(subset)})")
        continue
    rows_sub = _subset_pair_rows(subset, selected_layers_for_delta)
    by_layer = {}
    counts = {}
    for r in rows_sub:
        ln = r["layer"]
        counts[ln] = counts.get(ln, 0) + 1
        if ln not in by_layer:
            by_layer[ln] = {"cos_abs": 0.0, "spearman_chan": 0.0, "jaccard_topk": 0.0, "cka_abs": 0.0}
        for k in by_layer[ln].keys():
            by_layer[ln][k] += float(r[k])

    print(f"\nΔ similarity (mean over pairs) on selected layers within {tag}:")
    for ln in selected_layers_for_delta:
        if ln not in by_layer:
            continue
        c = counts[ln]
        m = {k: v / c for k, v in by_layer[ln].items()}
        print(f"- {ln}: cos={m['cos_abs']:.3f} spearman={m['spearman_chan']:.3f} jacc@k={m['jaccard_topk']:.3f} cka={m['cka_abs']:.3f} | n_pairs={c}")
```

## Δ-similarity: effect size (Cohen's d) и permutation p-value для success vs fail

Выше мы печатали **усреднённые** значения `cos/spearman/jacc/cka` внутри групп success и fail.
Теперь посчитаем статистику различий **по распределениям попарных значений** (внутри success-пар vs внутри fail-пар):
- `Cohen's d` для разности средних
- `permutation p-value` (как ниже в ноутбуке)

⚠️ Важно: значения по парам не независимы (одна и та же картинка входит во много пар), поэтому эти p-value нужно
трактовать как эвристику/индикатор силы эффекта, а не как строгий тест.


```python
# Compute pairwise rows within success and within fail once
rows_succ = compute_pairwise_delta_similarity(succ_items, selected_layers_for_delta, topk=CFG.topk_channels) if len(succ_items) >= 2 else []
rows_fail = compute_pairwise_delta_similarity(fail_items, selected_layers_for_delta, topk=CFG.topk_channels) if len(fail_items) >= 2 else []

metrics_delta = ["cos_abs", "spearman_chan", "jaccard_topk", "cka_abs"]

print("\nΔ-similarity stats (within success pairs vs within fail pairs):")
for ln in selected_layers_for_delta:
    rs = [r for r in rows_succ if r["layer"] == ln]
    rf = [r for r in rows_fail if r["layer"] == ln]
    if len(rs) == 0 or len(rf) == 0:
        print(f"- {ln}: not enough pairs (succ={len(rs)}, fail={len(rf)})")
        continue

    print(f"\nLayer: {ln} | n_pairs: succ={len(rs)} fail={len(rf)}")
    for m in metrics_delta:
        xs = np.array([float(r[m]) for r in rs], dtype=np.float64)
        xf = np.array([float(r[m]) for r in rf], dtype=np.float64)
        ms = float(np.mean(xs))
        mf = float(np.mean(xf))
        dval = cohens_d(xs, xf)
        pval = perm_pvalue_mean_diff(xs, xf, n_perm=5000, seed=7)
        print(f"  - {m}: mean_s={ms:.4f} mean_f={mf:.4f} d={dval:.3f} p={pval:.3f}")


attr_rows = []

for item in run_data:
    # ROI объекта: bbox топ-детекции на clean (прокси)
    obj_bbox = _get_target_bbox_from_results(item.get("res_clean", None), target_class_id)
    if obj_bbox is None:
        # если объекта нет — пропускаем, иначе ROI-метрики бессмысленны
        continue

    # подготовим clean input на устройство
    x_clean = torch_preprocess_letterboxed(item["clean_lb"]).to(CFG.device)

    for ln in roi_layers:
        delta = item["deltas"].get(ln, None)
        if delta is None:
            continue

        # MΔ
        md = spatial_energy(delta)  # [H,W]
        H, W = md.shape

        # MA через Grad*Act
        ch_imp, ma, ma_signed = grad_act_importance_maps(model_torch, x_clean, ln, target_class_id, obj_bbox_xyxy=obj_bbox)

        # ROI coords
        ox1, oy1, ox2, oy2 = _bbox_to_fm_coords(obj_bbox, H, W, CFG.imgsz)
        if ox2 <= ox1 or oy2 <= oy1:
            continue

        # ROI slices
        md_roi = md[oy1:oy2, ox1:ox2]
        ma_roi = ma[oy1:oy2, ox1:ox2]
        ma_signed_roi = ma_signed[oy1:oy2, ox1:ox2]

        # fractions
        ma_sum = _safe_sum(ma)
        ma_obj = _safe_sum(ma_roi)
        attr_obj_frac = (ma_obj / ma_sum) if ma_sum > 1e-12 else float("nan")

        # alignments
        cos_global = _cosine_map(md, ma)
        cos_roi = _cosine_map(md_roi, ma_roi)
        signed_roi_align = float(np.nanmean(md_roi * ma_signed_roi))
        signed_global_align = float(np.nanmean(md * ma_signed))

        attr_rows.append(
            {
                "path": item["path"],
                "success": bool(item["success"]),
                "drop": float(item["drop"]),
                "layer": ln,
                "attr_obj_frac": float(attr_obj_frac),
                "cos_global": float(cos_global),
                "cos_roi": float(cos_roi),
                "signed_roi_align": signed_roi_align,
                "signed_global_align": signed_global_align,
            }
        )

attr_df = pd.DataFrame(attr_rows)
print("rows:", len(attr_df))
attr_df

```

## Сравнение success/fail для ROI-gated метрик (effect size + permutation p)

Если гипотеза верна, ожидаем:
- `attr_obj_frac` выше для success
- `cos_roi` выше для success

`cos_global` может быть менее информативен, потому что успех/провал определяется тем, что происходит именно в области объекта.


```python
roi_metrics = ["attr_obj_frac", "cos_global", "cos_roi", "signed_roi_align", "signed_global_align"]

roi_summary_rows = []
for ln in roi_layers:
    dfl = attr_df[attr_df["layer"] == ln]
    if len(dfl) == 0:
        continue
    for m in roi_metrics:
        s = dfl[dfl["success"]][m].to_numpy(dtype=np.float64)
        f = dfl[~dfl["success"]][m].to_numpy(dtype=np.float64)
        roi_summary_rows.append(
            {
                "layer": ln,
                "metric": m,
                "mean_success": float(np.nanmean(s)) if np.any(~np.isnan(s)) else float("nan"),
                "mean_fail": float(np.nanmean(f)) if np.any(~np.isnan(f)) else float("nan"),
                "cohens_d": cohens_d(s, f),
                "perm_p": perm_pvalue_mean_diff(s, f, n_perm=3000, seed=1),
            }
        )

roi_summary = pd.DataFrame(roi_summary_rows)
roi_summary["abs_d"] = roi_summary["cohens_d"].abs()
print("Top ROI-gated signals:")
display(roi_summary.sort_values(["abs_d", "perm_p"], ascending=[False, True]).head(25))

```

## Агрегация ROI-gated метрик до уровня изображения

## Визуализация: карты энергии изменений (MΔ) и важности (MA) для success vs fail

 Мы сравниваем:
 - MΔ(h,w) = sum_c |Δ_l|  (где патч изменил признаки на слое)
 - MA(h,w) = sum_c |grad*act| (где признаки важны для ROI-conditioned скаляра)
 - MA_signed(h,w) = sum_c (grad*act) (знак/направление)

 Автоматически выбираем по одному success и одному fail примеру и один слой из roi_layers.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


def _norm01(m: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    m = m.astype(np.float32)
    mi = float(np.nanmin(m))
    ma = float(np.nanmax(m))
    if not np.isfinite(mi) or not np.isfinite(ma) or (ma - mi) < eps:
        return np.zeros_like(m, dtype=np.float32)
    return (m - mi) / (ma - mi)


def _get_first_example(run_data, success: bool):
    for it in run_data:
        if bool(it.get("success", False)) == bool(success):
            return it
    return None


def visualize_energy_and_importance(item, layer_name: str, title: str = ""):
    # ROI bbox объекта (на clean)
    obj_bbox = _get_target_bbox_from_results(item.get("res_clean", None), target_class_id)

    clean_img = item["clean_lb"]
    patched_img = item["patched_lb"]

    # Energy map from delta
    delta = item["deltas"].get(layer_name, None)
    if delta is None:
        print(f"No delta for layer={layer_name} on {item['path']}")
        return

    md = spatial_energy(delta)  # [H,W] = sum_c |Δ|
    H, W = md.shape

    # Importance maps via Grad*Act (abs + signed)
    x_clean = torch_preprocess_letterboxed(clean_img).to(CFG.device)
    ch_imp, ma_abs, ma_signed = grad_act_importance_maps(
        model_torch,
        x_clean,
        layer_name,
        target_class_id,
        obj_bbox_xyxy=obj_bbox,
    )

    # Map ROI bbox (pixel coords) to feature map coords
    roi_rect = None
    if obj_bbox is not None:
        ox1, oy1, ox2, oy2 = _bbox_to_fm_coords(obj_bbox, H, W, CFG.imgsz)
        if ox2 > ox1 and oy2 > oy1:
            roi_rect = (ox1, oy1, ox2 - ox1, oy2 - oy1)

    # Normalize for visualization
    md_n = _norm01(md)
    ma_n = _norm01(ma_abs)

    # Signed map: scale to [-1,1] by max abs
    s = ma_signed.astype(np.float32)
    smax = float(np.nanmax(np.abs(s))) if np.isfinite(np.nanmax(np.abs(s))) else 0.0
    s_vis = np.clip(s / (smax + 1e-12), -1.0, 1.0) if smax > 1e-12 else np.zeros_like(s, dtype=np.float32)

    fig, axes = plt.subplots(2, 3, figsize=(14, 8))

    # Clean image + bbox
    ax = axes[0, 0]
    ax.imshow(clean_img)
    ax.set_title(f"clean (lb)\n{title}")
    ax.axis("off")
    if obj_bbox is not None:
        x1, y1, x2, y2 = obj_bbox
        ax.add_patch(Rectangle((x1, y1), x2 - x1, y2 - y1, fill=False, linewidth=2))

    # Patched image + bbox
    ax = axes[0, 1]
    ax.imshow(patched_img)
    ax.set_title(f"patched (lb)\ndrop={item['drop']:.3f} success={item['success']}")
    ax.axis("off")
    if obj_bbox is not None:
        x1, y1, x2, y2 = obj_bbox
        ax.add_patch(Rectangle((x1, y1), x2 - x1, y2 - y1, fill=False, linewidth=2))

    # Overlay MΔ on clean image
    ax = axes[0, 2]
    ax.imshow(clean_img)
    ax.imshow(md_n, alpha=0.55, extent=(0, CFG.imgsz, CFG.imgsz, 0))
    ax.set_title(f"overlay: MΔ on clean\nlayer={layer_name}")
    ax.axis("off")

    # MΔ map
    ax = axes[1, 0]
    ax.imshow(md_n)
    ax.set_title("MΔ = sum_c |Δ| (norm)")
    ax.axis("off")
    if roi_rect is not None:
        ax.add_patch(Rectangle((roi_rect[0], roi_rect[1]), roi_rect[2], roi_rect[3], fill=False, linewidth=2))

    # MA abs map
    ax = axes[1, 1]
    ax.imshow(ma_n)
    ax.set_title("MA = sum_c |grad*act| (norm)")
    ax.axis("off")
    if roi_rect is not None:
        ax.add_patch(Rectangle((roi_rect[0], roi_rect[1]), roi_rect[2], roi_rect[3], fill=False, linewidth=2))

    # MA signed map
    ax = axes[1, 2]
    ax.imshow(s_vis, vmin=-1, vmax=1)
    ax.set_title("MA_signed = sum_c (grad*act) (scaled)")
    ax.axis("off")
    if roi_rect is not None:
        ax.add_patch(Rectangle((roi_rect[0], roi_rect[1]), roi_rect[2], roi_rect[3], fill=False, linewidth=2))

    plt.tight_layout()
    plt.show()


# Pick examples
ex_s = _get_first_example(run_data, success=True)
ex_f = _get_first_example(run_data, success=False)

# Choose layer
viz_layer = roi_layers[0] if ("roi_layers" in globals() and roi_layers) else layer_names[-2]
print("Visualization layer:", viz_layer)

if ex_s is not None:
    visualize_energy_and_importance(ex_s, viz_layer, title=f"SUCCESS: {ex_s['path']}")
else:
    print("No success example found.")

if ex_f is not None:
    visualize_energy_and_importance(ex_f, viz_layer, title=f"FAIL: {ex_f['path']}")
else:
    print("No fail example found.")
```

```python
## Heatmaps: попарная косинусная близость |Δ| (all / success-success / fail-fail / success-fail)

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from typing import Dict, Any, List, Optional, Sequence


def _delta_vec_abs(item: Dict[str, Any], layer: str) -> Optional[np.ndarray]:
    d = item.get("deltas", {}).get(layer, None)
    if d is None:
        return None
    return d.detach().cpu().abs().numpy().reshape(-1).astype(np.float64)


def _cos_vec(a: np.ndarray, b: np.ndarray, eps: float = 1e-12) -> float:
    na = np.linalg.norm(a)
    nb = np.linalg.norm(b)
    if na < eps or nb < eps:
        return float("nan")
    return float((a @ b) / (na * nb))


def cosine_matrix(items_a: Sequence[Dict[str, Any]], items_b: Sequence[Dict[str, Any]], layer: str) -> np.ndarray:
    Va = [_delta_vec_abs(it, layer) for it in items_a]
    Vb = [_delta_vec_abs(it, layer) for it in items_b]
    M = np.full((len(items_a), len(items_b)), np.nan, dtype=np.float64)
    for i, va in enumerate(Va):
        if va is None:
            continue
        for j, vb in enumerate(Vb):
            if vb is None:
                continue
            M[i, j] = _cos_vec(va, vb)
    return M


def show_heatmap(M: np.ndarray, title: str, vmin: float = 0.0, vmax: float = 1.0, annotate: bool = True, fmt: str = "{:.2f}"):
    # Bigger canvas for numeric annotations
    h, w = M.shape
    figsize = (max(7, 0.45 * w), max(6, 0.45 * h))
    plt.figure(figsize=figsize)

    im = plt.imshow(M, vmin=vmin, vmax=vmax)
    plt.colorbar(im)

    plt.title(title)
    plt.xlabel('j')
    plt.ylabel('i')

    if annotate:
        thresh = (vmin + vmax) / 2.0
        # Smaller font for large matrices
        fs = 8 if max(h, w) <= 12 else 6 if max(h, w) <= 20 else 4

        for i in range(h):
            for j in range(w):
                val = M[i, j]
                if not np.isfinite(val):
                    s = ""
                else:
                    s = fmt.format(float(val))
                # Contrast: dark text on bright cells, white text on dark
                color = "black" if (np.isfinite(val) and float(val) >= thresh) else "white"
                plt.text(j, i, s, ha="center", va="center", fontsize=fs, color=color)
        plt.tight_layout()
        plt.show()

# Слой: берём выбранный для Δ-анализа (если есть), иначе первый roi_layer, иначе предпоследний из layer_names
layer_for_cos = None
if "selected_layers_for_delta" in globals() and selected_layers_for_delta:
    layer_for_cos = selected_layers_for_delta[0]
elif "roi_layers" in globals() and roi_layers:
    layer_for_cos = roi_layers[0]
else:
    layer_for_cos = layer_names[-2]

print("Layer for cosine heatmaps:", layer_for_cos)

succ_items = [it for it in run_data if it.get("success", False)]
fail_items = [it for it in run_data if not it.get("success", False)]

# 1) all×all
M_all = cosine_matrix(run_data, run_data, layer_for_cos)
for i in range(M_all.shape[0]):
    if np.isfinite(M_all[i, i]):
        M_all[i, i] = 1.0
show_heatmap(M_all, f"cos(|Δ|) all×all (N={len(run_data)})\nlayer={layer_for_cos}", vmin=0.0, vmax=1.0)

# 2) success×success
if len(succ_items) >= 2:
    M_ss = cosine_matrix(succ_items, succ_items, layer_for_cos)
    for i in range(M_ss.shape[0]):
        if np.isfinite(M_ss[i, i]):
            M_ss[i, i] = 1.0
    show_heatmap(M_ss, f"cos(|Δ|) success×success (n={len(succ_items)})\nlayer={layer_for_cos}", vmin=0.0, vmax=1.0)
else:
    print("Not enough success items for success×success heatmap")

# 3) fail×fail
if len(fail_items) >= 2:
    M_ff = cosine_matrix(fail_items, fail_items, layer_for_cos)
    for i in range(M_ff.shape[0]):
        if np.isfinite(M_ff[i, i]):
            M_ff[i, i] = 1.0
    show_heatmap(M_ff, f"cos(|Δ|) fail×fail (n={len(fail_items)})\nlayer={layer_for_cos}", vmin=0.0, vmax=1.0)
else:
    print("Not enough fail items for fail×fail heatmap")

# 4) success×fail (прямоугольная)
if len(succ_items) >= 1 and len(fail_items) >= 1:
    M_sf = cosine_matrix(succ_items, fail_items, layer_for_cos)
    show_heatmap(M_sf, f"cos(|Δ|) success×fail ({len(succ_items)}×{len(fail_items)})\nlayer={layer_for_cos}", vmin=0.0,
                 vmax=1.0)
else:
    print("Not enough items for success×fail heatmap")
```
